```python
import cv2
import numpy as np
from keras.models import load_model
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
model = load_model('final_trained_model.h5')
```

```python
categories = ['airplane','bus','car','motorcycle','truck']
```

```python
video = cv2.VideoCapture(0)

while(True):
    succ,img = video.read()
    if not succ:
        print("End of Video")
        break
    rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(rgb_img,(150,150))
    img = cv2.flip(img,1)
    img_resized = img_resized/255.0
    img_reshaped = img_resized.reshape(1,150,150,3)
    obj = model.predict(img_reshaped)
    ind = np.argmax(obj)
    confidence = obj[0][ind]
    if confidence > 0.90:
        class_name = categories[ind]
    else:
        class_name = "unknown"
    text = f'{class_name} {confidence}' 
    cv2.putText(img, text, (30, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
    if cv2.waitKey(1) & 0XFF==113:
        break
    cv2.imshow('image',img)
 

video.release()
cv2.destroyAllWindows()
```

```python
def predict_new(pathfile,model1):
  categories = ['airplane','bus','car','motorcycle','truck']
  img = cv2.imread(pathfile)
  img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  resizeimg = cv2.resize(img_rgb,(150,150))
  resizeimg = resizeimg/255.0
  reshapedimg = resizeimg.reshape(1,150,150,3)
  predi = model1.predict(reshapedimg)
  inde = predi.argmax()
  return categories[inde]
```
